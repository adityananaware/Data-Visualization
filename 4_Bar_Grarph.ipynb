```python
#Bar Graph
import matplotlib.pyplot as plt
import numpy as np
```

```python
name = ['Bharat','Aditya','Siddhant','Vedant','Arush']
num = [4,2,7,5,1]
plt.bar(name,num,color = (['orangered','deepskyblue','pink','gold','m']))
#plt.barh(name,num,color = (['orangered','deepskyblue','pink','gold','m'])) - for horizontal graph
plt.title('Names And Numbers')
plt.xlabel('Names')
plt.ylabel('Numbers')
```

```python
x = np.array(['a','b','c','d'])
y = np.array([3,1,8,10])
plt.bar(x,y,color = 'pink', width = 0.1)
plt.title('Bar graph')
plt.xlabel('Names')
plt.ylabel('Numbers')
```
