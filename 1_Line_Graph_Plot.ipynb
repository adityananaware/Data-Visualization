```python
#Basic Colors
import matplotlib.pyplot as plt
import numpy as np
i = np.array([[[0,0,0]]])
plt.imshow(i)
```

```python
#Line graph/plot
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y, color = 'red', marker = 'x')
plt.title('LINE PLOT/GRAPH')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,color = 'orangered', marker = 'v')
plt.title('LINEGRAPH')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks(range(0,20,3))
plt.yticks(range(0,20,3))
```

```python
#Sport Watch Data
x = [60,70,80,90,100]
y = np.array([250,270,260,280,300])
plt.plot(x,y)
plt.title('Sport Watch Data')
plt.xlabel('Average Pluse')
plt.ylabel('Calories burned')
plt.grid()
```

```python
a = [1,2,3,4,5]
b = [2,6,1,3,5]
c = np.array([9,8,7,3,4])
plt.plot(a,b,linewidth = 5, linestyle = 'dashed')
plt.plot(a,c,linewidth = 2, linestyle = 'dotted')
plt.title('Linewidth and Linestyle')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```
