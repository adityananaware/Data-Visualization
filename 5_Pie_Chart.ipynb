```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([40,30,20,10])
mylabels = ['MI','Apple','Samsung','one Plus']
plt.pie(y,labels = mylabels)
```

```python
plt.pie(y,labels = mylabels, explode = [0,0,0,0.1])
```

```python
plt.pie(y,labels = mylabels, explode = [0.1,0.1,0.1,0.1])
```

```python
plt.pie(y,labels = mylabels, explode = [0.2,0,0,0])
```

```python
plt.pie(y,labels = mylabels, explode = [0.4,0.3,0.2,0.1])
```
